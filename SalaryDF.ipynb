```python
#Dependencies
import pandas as pd
import numpy as np

#Load file
salary_data = "Levels_Fyi_Salary_Data.csv/Levels_Fyi_Salary_Data.csv"

#Read into a dataframe
salary_data_df = pd.read_csv(salary_data)

#Check file
salary_data_df.head(2)
```

```python
Austin_salaries_df = salary_data_df[salary_data_df["location"]== "Austin, TX"]

Austin_salaries_df.tail(5)
```

```python
#Print all job titles for Austin based data
for val in Austin_salaries_df["title"].unique():
    print(val)
```

```python
# Calculate the Total Companies in Austin
Austin_companies = len(Austin_salaries_df["company"].unique())
Austin_count = Austin_salaries_df["timestamp"].count()

# Calculate average salary for Austin
Austin_avg_salary = round(Austin_salaries_df["totalyearlycompensation"].mean(), 2)
```

```python
f"This dataset contains {Austin_companies} companies in Austin, TX with a total of {Austin_count} salaries reported. The average salary of tech jobs in Austin is ${Austin_avg_salary}."
```
